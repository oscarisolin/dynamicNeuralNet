# Stabilit채t
* Gradientenabstiegverfahren scheint weniger anf채llig f체r oszillation, ist aber auch langsamer in der Konvergenz. 
  Die Korrektur verkleinert sich bei falchen kurven / kelinen Gradienten.
* Gauss-Newton Verfahren ist schneller, neigt aber zu Oszillation. vielleicht ist das bei Neuronalen Netzen nicht so erw체nscht.

```python
from matplotlib import pyplot as plt
import numpy as np

def f(x):
    return np.sin(x)

def df(x,step=0.0001):
    return np.divide(f(x+step)-f(x),step)
    

stellen = np.arange(0,20,np.pi*0.01)

plt.plot(stellen,f(stellen))
old_val = 20
old_val_gauss = old_val
plt.plot(old_val,f(old_val),'o',color='black')
for a in range(1,6):
    dz = 0-f(old_val_gauss)
    new_val = old_val - 0.1*df(old_val)
    new_val_gauss = old_val_gauss + dz/df(old_val_gauss)
    plt.plot(new_val,f(new_val),'o', color='red')
    plt.text(new_val, f(new_val), str(a), color="red", fontsize=12)
    old_val = new_val
    plt.plot(new_val_gauss,f(new_val_gauss),'o', color='blue')
    plt.text(new_val_gauss, f(new_val_gauss), str(a), color="blue", fontsize=12)
    old_val_gauss = new_val_gauss
plt.show()

```
